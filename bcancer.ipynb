```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()

df = pd.DataFrame(data.data,columns=data.feature_names)
df['target'] = data.target
```

```python
y = df['target']
x = df.drop('target', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=30,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, ConfusionMatrixDisplay
from sklearn.metrics import  recall_score, f1_score, roc_auc_score, confusion_matrix
```

```python
accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1_score = f1_score(y_test,y_pred)
roc_auc_score = roc_auc_score(y_test,y_pred)
confusion = confusion_matrix(y_test,y_pred)
```

```python
print(accuracy)
print(precision)
print(recall)
print(f1_score)
print(roc_auc_score)
matrix = ConfusionMatrixDisplay(confusion, display_labels= model.classes_)
matrix.plot()
```
